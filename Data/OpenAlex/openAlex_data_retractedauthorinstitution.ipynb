```python
def get_retracted_authors_institutions(pager,max_limit_pages):
  # Parameter: pager           -> OpenAlex object
  #            max_limit_pages -> How many pages to retrieve (set to -1 for all pages)

  retractedwork_author_pairs = pd.DataFrame(
      columns=[
        'workid',
        'authorid',
        'displayname'
    ]
  )
  retractedwork_institution_pairs = pd.DataFrame(
      columns=[
        'workid',
        'institutionid',
        'displayname'
    ]
  )

  page_count = 0
  for page in pager:
    print(f"Page: {page_count}/{max_limit_pages}")

    retractedwork_author_pairs = pd.concat([
      retractedwork_author_pairs,
      pd.DataFrame([
          (work['id'],author['author']['id'],author['author']['display_name'])
              for work in page
                  for author in work['authorships']
          ],
          columns=[
              'workid',
              'authorid',
              'displayname'
          ]
      )
    ])

    retractedwork_institution_pairs_to_append = pd.DataFrame([
        (work['id'],institution['id'],institution['display_name'])
            for work in page
                for author in work['authorships']
                    for institution in author['institutions']
        ],
        columns=[
            'workid',
            'institutionid',
            'displayname'
        ]
    )
    retractedwork_institution_pairs_to_append.drop_duplicates(inplace=True)
    retractedwork_institution_pairs = pd.concat([
      retractedwork_institution_pairs,
      retractedwork_institution_pairs_to_append
    ])

    page_count += 1
    if page_count == max_limit_pages:
      break

  return retractedwork_author_pairs, retractedwork_institution_pairs
```

```python
pager = Works().filter(
    is_retracted=True,
    publication_year=">2022"
    ).paginate(per_page=200)
```

```python
retractedwork_author_pairs_example, \
  retractedwork_institution_pairs_example = \
    get_retracted_authors_institutions(pager,-1)
```

```python
print(len(retractedwork_author_pairs_example))
print(len(retractedwork_institution_pairs_example))
```

```python
# Summary of authors
retractedwork_author_pairs_example['displayname'].value_counts()
```

```python
# Summary of institutions
retractedwork_institution_pairs_example['displayname'].value_counts()
```
