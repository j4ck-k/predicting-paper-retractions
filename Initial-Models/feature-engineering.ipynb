```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yake
```

```python
filename = './Data/retractions.csv'

raw_data = pd.read_csv(
    filename,
    encoding='unicode_escape'
)

raw_data.head()
```

```python
language = "en"
max_ngram_size = 2
deduplication_threshold = 0.9
numOfKeywords = 3
kwe = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, top=numOfKeywords, features=None)
extract_kw = lambda x : [k[0] for k in kwe.extract_keywords(x)]
```

```python
raw_data['Keywords'] = raw_data['Title'].apply(extract_kw)
```

```python
kw_split = pd.DataFrame(raw_data['Keywords'].to_list(), columns = ['TopKeyword', 'SecondKeyword', 'ThirdKeyword'])
```

```python
data = pd.concat([raw_data, kw_split], axis=1) 

data.head()
```

```python
count_col = lambda x : len(x.split(';'))
```

```python
data['NumAuthors'] = data['Author'].apply(count_col)
data['NumCountries'] = data['Country'].apply(count_col)
data['NumReasons'] = data['Reason'].apply(count_col)-1
```

```python
data.sort_values(by='NumAuthors')
```

```python
remove_time  = lambda x : x.split()[0]
```

```python
data['RetractionDate'] = pd.to_datetime(raw_data['RetractionDate'].apply(remove_time))
data['OriginalPaperDate'] = pd.to_datetime(raw_data['OriginalPaperDate'].apply(remove_time))
data['RetractionDt'] = (data['RetractionDate'] - data['OriginalPaperDate']).dt.days
data['RetractionDtYear'] = data['RetractionDt']/365
data[['RetractionDate', 'OriginalPaperDate', 'RetractionDt', 'RetractionDtYear']]
```
